```python
import numpy as np
import matplotlib.pyplot as plt
import random

def norm_list(a_l):
    s = sum(a_l)
    for i in range(len(a_l)):
        a_l[i] /= s
    return a_l

# 准备数据
x_data = ["NYC","TKY","IST","JK","KL","SP"]
y1_data = [0.119,0.171,0.053,0.069,0.114,0.154]
y2_data = [26.33,96.70,87.60,59.20,81.84,63.19] #签到
y3_data = []

y1_data = norm_list(y1_data)
y2_data = norm_list(y2_data)

for i in range(len(y1_data)):
    y3_data.append(y2_data[i]/y1_data[i])
    
y3_data = norm_list(y3_data)
    
# 正确显示中文和负号
fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# 画图，plt.bar()可以画柱状图
for i in range(len(x_data)):
    axes[0].bar(x_data[i], y1_data[i])
    axes[1].bar(x_data[i], y2_data[i])
    axes[2].bar(x_data[i], y3_data[i])
    
    axes[0].set_xticks([i for i in range(len(x_data))], x_data, fontproperties='Times New Roman', size=13)
    axes[1].set_xticks([i for i in range(len(x_data))], x_data, fontproperties='Times New Roman', size=13)
    axes[2].set_xticks([i for i in range(len(x_data))], x_data, fontproperties='Times New Roman', size=13)
    
    axes[0].set_xlabel('a.Social Sparsity Comparison', fontproperties='Times New Roman', size=18)  # 添加横轴标签
    axes[0].set_ylabel("D^s", fontproperties='Times New Roman',  size=20)  # 添加纵轴标签
    axes[0].grid(True)
    
    axes[1].set_xlabel('b.Mobile Sparsity Comparison', fontproperties='Times New Roman',  size=18)  # 添加横轴标签
    axes[1].set_ylabel("D^m", fontproperties='Times New Roman',  size=20)  # 添加纵轴标签
    axes[1].grid(True)
    
    axes[2].set_xlabel('c.The ratio of mobile and social sparsity', fontproperties='Times New Roman',  size=18)  # 添加横轴标签
    axes[2].set_ylabel("D^sm", fontproperties='Times New Roman',  size=20)  # 添加纵轴标签
    axes[2].grid(True)
# 显示
fig.legend(x_data, loc='upper center', ncol=len(x_data), bbox_to_anchor=(0.5, 1.02),  fontsize=20)  # 绘制前len(methods)个图例 
plt.savefig('result_pic/Density_comparison_result.jpg', dpi=200, bbox_inches='tight')
plt.show()
```
